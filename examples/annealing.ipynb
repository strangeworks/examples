# Annealing example
In this example we will use the `strangeworks-annealing` package to solve the 5-node max-cut problem. The problem formulated as a QUBO problem and submitted to a quantum annealer. To install the package, run `pip install strangeworks-annealing`. For more information, see the [documentation](https://docs.strangeworks.com/sdk-documentation/frameworks-extensions/annealing).

````python
import strangeworks
import strangeworks.annealing
````

Authenticate to Strangeworks

````python
strangeworks.authenticate(user, key)
````


Fetch available annealing backends


````python
print(strangeworks.annealing.get_solvers())
````


Select a backend. You can use any listed from the previous step. We will use `aws.DW_200Q_6` for this example.

````python
solver = strangeworks.annealing.select_solver("aws.DW_2000Q_6")
````


You can upload a qubo problem from a file by using the `upload_qubo` method.
```
problem = upload_problem("./maxcut.qubo")
```

Alternatively create a problem by setting the linear and quadratic coefficients directly as we do here.



````python
linear = {1: -2, 2: -2, 3: -3, 4: -3, 5: -2}
quadratic = {(1, 2): 2, (1, 3): 2, (2, 4): 2, (3, 4): 2, (3, 5): 2, (4, 5): 2}
problem = strangeworks.annealing.problem.Problem(linear, quadratic)
````


Run the job, unembed the result, and print the output. This action blocks until a result is fetched from Strangeworks


````python
job = solver.optimize(problem=problem, shots=10)
````

````python
print(solver.unembed_result(job))
````
