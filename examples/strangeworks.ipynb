```python
from qiskit import QuantumCircuit

from strangeworks.qiskit.provider import StrangeworksProvider

provider = StrangeworksProvider()
# the backend name comes from provider.backends()
backend = provider.get_backend("ibm.strangeworks-ibm.ibmq_qasm_simulator")
# backend = provider.get_backend("ibm.strangeworks-ibm.ibm_nairobi")

qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])


job = qiskit.execute(qc, backend, shots=50)
print(f"id: {job.job_id()}, status: {job.status()}")
# But you can also print your results manually as well:
result = job.result()
print("result:", result)
counts = result.get_counts()
print("counts:", counts)
```
