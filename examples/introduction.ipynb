### StrangeworksProvider

We implement our own `qiskit.providers.provider.ProviderV1`, called `StrangeworksProvider`.
`StrangeworksProvider` gives you the expected functionality of a qiskit.ProviderV1.

You can determine which backends are available to you through our platform. Check it out.

```python
from strangeworks.qiskit import StrangeworksProvider
provider = StrangeworksProvider()

provider.backends()
```

### Backend Name
As you can see above, I have multiple available backends.
When requesting a backend from our provider, you must utilize one of the names listed above.
The backend name usually follows the convention of provider-id.your-account-unique-code-id.backend-name.

`your-account-unique-code-id` is the UNIQUE CODE ID displayed in your accounts page. For example you can view your IBM accounts [here](https://app.quantumcomputing.com/hardware/ibm/accounts).

Let me request a Strangeworks managed IBM backend.

```python
santi = provider.get_backend("ibm.strangeworks-ibm.ibmq_santiago")

name = santi.name()
status = santi.status()
properties = santi.properties()
print(f"{name} status: {status.to_dict()}")
if properties:
    print(f"{name} properties: \n{properties.to_dict()}")
```

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])

qasm_sim = provider.get_backend("ibm.strangeworks-ibm.ibmq_qasm_simulator")
job = qasm_sim.run(qc, shots=50)
# job_id() corresponds to the job name displayed here https://app.quantumcomputing.com/jobs
print(f"{job.job_id()}")
```
