```python
from qiskit import QuantumCircuit, execute
from strangeworks.qiskit.provider import StrangeworksProvider
```

```python
provider = StrangeworksProvider()
```

```python
# List available backends
provider.backends()
```

```python
# utilize Rigetti via AWS
# Make sure backend is in the list above
backend = provider.get_backend("aws.aws-managed.Aspen-11")
```

```python
backend.status()
```

```python
# reset result id
# strangeworks.client.result_id = ""
```

```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])
qc.draw()
```

```python
# execute a simple circuit
job = backend.run(qc, shots = 10)
print(job.job_id())
```

```python
print(f"https://app.quantumcomputing.com/results/{strangeworks.client.result_id}")
```

```python
from strangeworks.qiskit.jobs.strangeworksjob import StrangeworksJob

# we can fetch old job results too, update with job ID from above
j = StrangeworksJob.get(job_id=job.job_id())
print(j.result().get_counts())
```
